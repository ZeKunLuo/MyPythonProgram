```python
import requests
import urllib.request
import time#å¯¼å¥å
r=requests.get('http://www.image-net.org/api/text/imagenet.synset.geturls?wnid=n04105893')
names=r.text
url=names.split('\r\n')
imgname=[]
for i in range(0,len(url)):
    try:
        print ( url[i]+"->"+str(i+1)+'.jpg')
        imgname.append('E:/Download/image2/'+str(i+1)+'.jpg')
        web = urllib.request.urlopen(url[i])
        #time.sleep(3)#è®¾ç½®æ¶é´é´éä¸º3ç§
        data = web.read()
        f = open(imgname[i],"wb")
        f.write(data)
        f.close()
        print ("->downloaded!")
    except urllib.request.HTTPError as reason:
        print(",fail to download, caused by "+reason.reason)
        continue
    except urllib.request.URLError as reason:
        print(",fail to download, caused by URLError")
        continue
    except:
        print(", FAILED for unknow reason!")
        continue
```
